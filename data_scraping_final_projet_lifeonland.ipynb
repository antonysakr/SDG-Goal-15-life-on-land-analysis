```python
 import requests
from bs4 import BeautifulSoup
url = 'https://en.wikipedia.org/wiki/Life_on_Land'
r = requests.get(url)
html_content = r.text
html_soup = BeautifulSoup(html_content,"html.parser")
```

```python
#beautiful soup
print("title : ")
print(html_soup.find("title"))

print("*"*40)
print("number of paragraphs in the url : ")
print(len(html_soup.find_all("p")))

print("*"*40)
print("text in fifth p and it's length:")
print(html_soup.find_all("p")[4].text)
print(len(html_soup.find_all("p")[4].text))

print("*"*40)
print("first three h2 tags : ")
print(html_soup.find_all("h2")[0:3])

print("*"*40)
print("href of the fourth /a tag : ")
print(html_soup.find_all("a")[3].attrs['href'])

print("*"*40)
print("list of all h1 , h2, h3 :")
for h in html_soup.find_all(["h1","h2","h3"]):
    print(h.name + " " + h.text.strip())

print("*"*40)
print("childrens of the body tag : ")
childs = [x.name for x in html_soup.body.children if x.name is not None ]
print(childs)

print("*"*40)
print("All li tags: ")
for tag in html_soup.find_all("li"):
    print("{a}: {b}".format(a=tag.name, b=tag.text.strip()))
```

```python
#regular expression
import re
import datetime
from datetime import date

y= re.findall("^the",r.text)
if y:
    print("the text starts with 'the'")
else:
        print("the text does not starts with 'the'")

print("*"*40)  
print("first 10 digits : ")
print(re.findall("\d",r.text)[0:10]) 

print("*"*40) 
print("all words that start with 'tr' then exactly two 'e' : ")
print(re.findall("tre{2}",r.text))

print("*"*40)
x = re.search("\s", r.text)
print("The first white-space character is located in position:", x.start())

print("*"*40)
print("extract datesfrom the text : ")
match = re.search(r'\d{4}-\d{2}-\d{2}', r.text)
date = datetime.datetime.strptime(match.group(), '%Y-%m-%d').date()
print (match)

print("*"*40)
print("split the text in the first 500 white spaces : ")
print(re.split("\s", r.text,500))

print("*"*40)
print("replace all digits with a '$'")
print(re.sub("\d", "$", r.text))

print("*"*40)
print("position of the first word starts with 'S' : ")
x = re.search(r"\bS\w+", r.text)
print(x.span())

print("*"*40)
print("return the string that contain 'S' : ")
y = re.search(r"\bS\w+", r.text)
print(y.string)

print("*"*40)
z = re.search(r"\bS\w+", r.text)
print(z.group())
```

```python
#web crawler
from urllib.request import urlopen
from urllib.parse import urlparse
from bs4 import BeautifulSoup
import re
import datetime
import random
pages=set()
random.seed(datetime.datetime.now())
html=urlopen('https://en.wikipedia.org/wiki/Sustainable_Development_Goal_15')
bs=BeautifulSoup(html,'html.parser')
#Retrieves a list of all Internal linksfound on a page
def getInternalLinks(bs,includeUrl):
    includeUrl='{}://{}'.format(urlparse(includeUrl).scheme,urlparse(includeUrl).netloc)
    internalLinks=[]
    #Finds all links that begin with a "/"
    for link in bs.find_all('a',href=re.compile('^(/|.*'+includeUrl+')')):
        if link.attrs['href'] is not None:
            if link.attrs['href'] not in internalLinks:
                if(link.attrs['href'].startswith('/')):
                    internalLinks.append(includeUrl+link.attrs['href'])
                else:
                    internalLinks.append(link.attrs['href'])
                    return internalLinks
                #....

getInternalLinks(bs,'https://en.wikipedia.org/wiki/Sustainable_Development_Goal_15')
```
